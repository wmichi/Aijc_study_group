```python
pre_columns =  app_train.select_dtypes('object').apply(pd.Series.nunique, axis = 0)
pre_columns.index
# 後でonehot用に残しておく
```

# 前処理

```python
app_train.shape
```

```python
ID = app_train['SK_ID_CURR']
# X =  X.drop(ID,axis=1)
y = app_train['TARGET']
X = app_train.drop('SK_ID_CURR',axis=1)
X = X.drop('TARGET',axis=1)
X.head()
```

```python
print( y.value_counts)
```

```python
ohe_columns = pre_columns.index
X_new = pd.get_dummies(X,
                      dummy_na=True,
                      columns=ohe_columns)
X_new.head()
```

```python
X_new.describe()
```

```python
# つまりnull値はないってこと?
X_new[X_new.isnull().all(axis=1)]
print(X_new.shape)
```

```python
from sklearn.impute import SimpleImputer # impute で欠損値を入れる
imp = SimpleImputer() # 欠損値に平均値を入れる。
imp.fit(X_new)
X_new_columns =  X_new.columns.values
X_new = pd.DataFrame(imp.transform(X_new),columns=X_new_columns)
display(X_new.head())
```

```python
from sklearn.preprocessing import MinMaxScaler
mmsc = MinMaxScaler()
X_new_norm = mmsc.fit_transform(X_new)
```

```python
from sklearn.feature_selection import RFE
from sklearn.ensemble import RandomForestClassifier
selector =  RFE(estimator= RandomForestClassifier(n_estimators= 100,random_state= 0),
               n_features_to_select=30,
               step= 0.5)
selector.fit( X_new,y)
print('Done normally')
```
